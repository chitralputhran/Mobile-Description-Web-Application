**Setting up the dataframe**

```python
device_brands = ['apple', 'samsung', 'sony', 'huawei', 'nokia', 'honor', 'lg', 'htc', 'motorola', 'blackberry']
max_features = 0
for brand in device_brands:
    phones = fon.getdevice(brand)
    for phone in phones:
        if len(phone.keys()) > max_features:
            max_features = len(phone.keys())
            features = phone.keys()
```

```python
data = pd.DataFrame(data = '', columns=features , index = range(883))
```

```python
count = 0
device_brands = ['apple', 'samsung', 'sony', 'huawei', 'nokia', 'honor', 'lg', 'htc', 'motorola', 'blackberry']
for brand in device_brands:
    phones = fon.getdevice(brand)
    for phone in phones:
        if (type(phone) == type(dict())):
            for column_name in data.columns:
                if column_name in phone.keys():
                    data.loc[count][column_name] = phone[column_name]
                else:
                    data.loc[count][column_name] = np.nan
            
            count = count + 1


```
